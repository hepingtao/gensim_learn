```python
#coding: utf-8
import sys, jieba
from gensim import corpora, models, similarities
# reload(sys)
# sys.setdefaultencoding('utf8')

texts = [
    '什么是股票？',
    '股票是个什么玩意？',
    '新手怎样入门炒股？',
    '现在股市的风险大吗？',
    'python的自然语言处理',
    'gensim的主要功能有把文本转为向量',
    '提供存储文本矩阵的方法',
    '这篇文章主要讨论如何用python来做一些简单的文本处理'
]
docs = [jieba.lcut_for_search(i) for i in texts]

dic = corpora.Dictionary(docs)
corpus = [dic.doc2bow(i) for i in docs]

lsi = models.LsiModel(corpus, id2word=dic, num_topics=2)

qurey1 = "股票"
qurey2 = "文本"
vec_query1 = dic.doc2bow([qurey1])
vec_query2 = dic.doc2bow([qurey2])

index = similarities.MatrixSimilarity(lsi[corpus])

sims1 = index[lsi[vec_query1]]
sims2 = index[lsi[vec_query2]]
sims_result1 = sorted(enumerate(sims1), key=lambda item: -item[1])
sims_result2 = sorted(enumerate(sims2), key=lambda item: -item[1])
print(sims_result1)
print(sims_result2)
```

```python
class MySqlDB():
    def __init__(self, **kwargs):
        self.dbconn = None
        self.dbcursor = None
        self.conn_db(**kwargs)

    def conn_db(self, **kwargs):
        print(kwargs)
        charset = kwargs['charset']
        dbconf = kwargs['dbconf']
        host = dbconf['host']
        user = dbconf['user']
        passwd = dbconf['passwd']
        dbname = dbconf['dbname']
        port = dbconf['port']

        try:
            self.dbconn = pymysql.connect(host=host, user=user, passwd=passwd, db=dbname, port=port, charset=charset)
            self.dbconn.autocommit(True)
            self.dbconn.ping(True)
            self.dbcursor = self.dbconn.cursor(pymysql.cursors.Cursor)
        except Exception:
            raise Exception("conn mysql error")

    def execute_sql(self, sql):
        print(sql)
        try:
            self.dbcursor.execute(sql)
        except Exception:
            if e[0] != 1062:
                raise Exception("execute mysql error")

    def executemany_sql(self, sql, data):
        self.dbcursor.executemany(sql, data)

    def get_data(self):
        return self.dbcursor.fetchall()

    def query_with_sql(self, sql):
        self.execute_sql(sql)
        data = self.get_data()
        return data

    def get_user_tags(self, root_tag_pk):
        return self.query_with_sql(get_user_tags_sql.format(root_tag_pk=root_tag_pk))

    def get_channels(self, channel_pks=[]):
        channels = self.query_with_sql(get_channels_sql)
        if channel_pks:
            channels = [channel for channel in channels if channel[1] in channel_pks]
        return channels

    def get_news(self, channel_pk, last_max_news_pk, limit_count, news_pks=[]):
        return self.query_with_sql(get_news_sql.format(channel_pk=channel_pk,
                                                       last_max_news_pk=last_max_news_pk,
                                                       limit_count=limit_count,
                                                       news_pks=",".join([str(x) for x in news_pks])))

    def write_with_sql(self, sql):
        self.execute_sql(sql)

    def batch_read(self, sql, data):
        self.executemany_sql(sql, data)
        data = self.get_data()
        return data

    def batch_write(self, sql, data, sub_batch_num=None):
        if sub_batch_num and sub_batch_num <= 0:
            return
        start = 0
        end = sub_batch_num
        while data[start:end]:
            self.executemany_sql(sql, data[start:end])
            if sub_batch_num:
                start, end = end, end + sub_batch_num
            else:
                start, end = 0, 0

    def batch_write_keywords(self, data):
        self.batch_write(batch_write_keywords_sql, data, sub_batch_num=batch_write_keywords_sub_batch_num)

    def close_db(self):
        try:
            self.dbconn.close()
            self.dbcursor.close()
        except Exception:
            raise Exception("close mysql error")
            # g_logger.error(traceback.format_exc())


class MySQLDev(MySqlDB):
    def __init__(self):
        self.conn_db(dbconf=mysql_conf_dev, charset="utf8")


class MySQLTest(MySqlDB):
    def __init__(self):
        self.conn_db(dbconf=mysql_conf_test, charset="utf8")


class MySQLProd(MySqlDB):
    def __init__(self):
        self.conn_db(dbconf=mysql_conf_prod, charset="utf8")


class MySQL(MySqlDB):
    def __init__(self, env='dev'):
        if env == 'dev':
            self.conn_db(dbconf=mysql_conf_dev, charset="utf8")
        if env == 'test':
            self.conn_db(dbconf=mysql_conf_test, charset="utf8")
        if env == 'prod':
            self.conn_db(dbconf=mysql_conf_prod, charset="utf8")
```
